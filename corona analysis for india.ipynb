# Corona positive cases in each states of India

```python
datasets.groupby(['Name of State / UT'])['Total Confirmed cases'].max().sort_values().plot(kind="bar",figsize=(15,8), color='b', title="Covid-19 Positive cases in India")
```

```python
print("corona cured cases in each states")
datasets.groupby(['Name of State / UT'])['Cured/Discharged/Migrated'].max().sort_values()
```

# Corona Cured cases in each states of India

```python
datasets.groupby(['Name of State / UT'])['Cured/Discharged/Migrated'].max().sort_values().plot(kind="bar",figsize=(14,8),color='g',title="Recovery Rate of Covid-19 in India")
```

```python
print("Corona death cases in each states")
datasets.groupby(['Name of State / UT'])['Death'].max().sort_values()
```

# Corona Death cases in each states of India

```python
datasets.groupby(['Name of State / UT'])['Death'].max().sort_values().plot(kind="bar",figsize=(15,9),title="Death rate of covid-19 in India",color='r')
```

```python
print("Corona confirmed cases in each days from 30/01/2020 to 08/04/2020")
datasets.groupby(['Date'])['Total Confirmed cases'].sum()
```

# Growth Rate of Covid-19 Positive Cases in India

```python
datasets.groupby(['Date'])['Total Confirmed cases'].sum().plot(kind="line",figsize=(15,9),title="Growth Rate of Covid-19 Positive Cases in India")
```

```python
print("Corona death cases in each days from 30/01/2020 to 08/04/2020")
datasets.groupby(['Date'])['Death'].sum()
```

# Death Rate of Covid-19 Cases in India

```python
datasets.groupby(['Date'])['Death'].sum().plot(kind="line",figsize=(15,9),title="Death Rate of Covid-19 Cases in India",color='r')
```

```python
print("Corona cured cases in each days from 30/01/2020 to 08/04/2020")
datasets.groupby(['Date'])['Cured/Discharged/Migrated'].sum()
```

# Recovery Rate of Covid-19 Cases in India

```python
datasets.groupby(['Date'])['Cured/Discharged/Migrated'].sum().plot(kind="line",figsize=(15,9),title="Recovery Rate of Covid-19 Cases in India",color='g')
```

```python
style.use('ggplot')
plt.figure(figsize=(20,9))
plt.plot(datasets.groupby(['Date'])['Total Confirmed cases'].sum(),'b', label="Corona positive cases", linewidth=4)
plt.plot(datasets.groupby(['Date'])['Death'].sum(),'r', label="corona death cases", linewidth=5)
plt.plot(datasets.groupby(['Date'])['Cured/Discharged/Migrated'].sum(),'g', label="corona cured cases", linewidth=6)
plt.title("Covid-19 cases in India")
plt.legend()
plt.show()
```

```python
tamilnadu=datasets.loc[datasets['Name of State / UT']=='Tamil Nadu']
tamilnadu.head()
```

# Corona update in Tamil Nadu till 08/04/2020

```python
style.use('ggplot')
plt.figure(figsize=(15,9))
plt.plot(tamilnadu.groupby(['Date'])['Total Confirmed cases'].sum(),'b', label="Corona positive cases", linewidth=4)
plt.plot(tamilnadu.groupby(['Date'])['Death'].sum(),'r', label="corona death cases", linewidth=5)
plt.plot(tamilnadu.groupby(['Date'])['Cured/Discharged/Migrated'].sum(),'g', label="corona cured cases", linewidth=6)
plt.title("Covid-19 cases in Tamilnadu")
plt.legend()
plt.show()
```
